```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
```

```python
# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
# invert access to show Dec on bottom
plt.gca().invert_yaxis()
plt.ylabel("Date")
plt.xlabel("Fare($)")
plt.legend()

```

```python
plt.scatter(y_axis,x_axis, color="red", linewidth=2, label='Chicago')
# Create labels for the x and y axes.
plt.xlabel("Month")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(-1, 12)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend
plt.legend()
plt.gca().invert_yaxis()
```

```python
y_axis_larger=[]
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# list comprehension instead of for and while loops
plt.scatter(x_axis, y_axis, s= [i*3 for i in y_axis])
```

```python
fig, ax=plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], c='xkcd:sky blue', alpha=.8, edgecolor='k', linewidth=2, label="Boston")


ax.set_title("PyBer Fare by Month")
ax.legend(labels ="Boston",loc='upper right')

ax.set_ylim(-1,12)
ax.set_xlim (0,60)
ax.set_xlabel ("Fares ($)")
ax.set_ylabel ("Months")
plt.gca().invert_yaxis()
```

```python
# Import dependencies.
import matplotlib.pyplot as plt
import statistics
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Get the standard deviation of the values in the y-axis.
stdev = statistics.stdev(y_axis)
stdev
```

```python
plt.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```

```python
plt.bar(x_axis, y_axis, yerr=stdev, capsize=3)
```

```python
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=stdev, capsize=3)
plt.show()
```
